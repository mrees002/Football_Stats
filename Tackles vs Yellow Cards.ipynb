```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
PremierLeague = pd.read_csv("PL Tackle Stats - Sheet1.csv")
LaLiga = pd.read_csv("La Liga Miscellaneous Stats - Sheet1.csv")
SerieA = pd.read_csv("Serie A Miscellaneous Stats - Sheet1.csv")
Bundesliga = pd.read_csv("Bundesliga Miscellaneous Stats - Sheet1.csv")
LigueUn = pd.read_csv("Ligue Un Miscellaneous Stats - Sheet1.csv")
```

```python
PremierLeague_Yellows = list(PremierLeague['CrdY'])
PremierLeague_Tackles = list(PremierLeague['TklW'])
PremierLeague_Players = list(PremierLeague['Player'])

LaLiga_Yellows = list(LaLiga['CrdY'])
LaLiga_Tackles = list(LaLiga['TklW'])
LaLiga_Players = list(LaLiga['Player'])

SerieA_Yellows = list(SerieA['CrdY'])
SerieA_Tackles = list(SerieA['TklW'])
SerieA_Players = list(SerieA['Player'])

Bundesliga_Yellows = list(Bundesliga['CrdY'])
Bundesliga_Tackles = list(Bundesliga['TklW'])
Bundesliga_Players = list(Bundesliga['Player'])

LigueUn_Yellows = list(LigueUn['CrdY'])
LigueUn_Tackles = list(LigueUn['TklW'])
LigueUn_Players = list(LigueUn['Player'])
```

```python
plt.scatter(PremierLeague_Tackles,PremierLeague_Yellows)
plt.xlabel('Tackles that won Possession per 90 minutes')
plt.ylabel('Yellow Cards per 90 minutes')
for i in range(len(PremierLeague_Players)):
    if (PremierLeague_Yellows[i]> 0.5 and PremierLeague_Tackles[i] > 2.0):
        plt.annotate(PremierLeague_Players[i],(PremierLeague_Tackles[i], PremierLeague_Yellows[i]))
    elif (PremierLeague_Yellows[i]< 0.5 and PremierLeague_Tackles[i] > 2.55):
        plt.annotate(PremierLeague_Players[i],(PremierLeague_Tackles[i], PremierLeague_Yellows[i]))

plt.annotate(PremierLeague_Players[70],(PremierLeague_Tackles[70], PremierLeague_Yellows[70]))
plt.annotate(PremierLeague_Players[164],(PremierLeague_Tackles[164], PremierLeague_Yellows[164]))
```
