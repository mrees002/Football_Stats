```python
import pandas as pd
import matplotlib.pyplot as plt
import mplcyberpunk
```

```python
df = pd.read_csv('EPL goals vs age - Sheet1.csv')
df[['Years', 'Days']]= df['Age'].str.split('-', expand = True)
df.head()
```

```python
plt.style.use('cyberpunk')
plt.scatter(x = 'Born', y = 'Gls', data = df, c = 'xGp90', vmin = 0, vmax = 1)
plt.xlabel('Year Born')
plt.ylabel('Goals')
plt.colorbar(label = 'xgp90')
mplcyberpunk.make_scatter_glow()
```
