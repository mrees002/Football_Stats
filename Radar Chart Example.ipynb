```python
import pandas as pd
import matplotlib.pyplot as plt
from soccerplots.radar_chart import Radar
```

```python
df = pd.read_csv('Fab vs Casemiro - Sheet1.csv')
```

```python
df
```

```python
params = list(df.columns)
params = params[1:]
```

```python
ranges = []
a_values = []
b_values = []

for x in params:
    a = min(df[params][x])
    a = a - (a*0.25)
    
    b = max(df[params][x])
    b = b + (b*0.25)
    
    ranges.append((a,b))
    
for x in range(len(df['Player'])):
    if df['Player'][x] == 'Fabinho':
        a_values = df.iloc[x].values.tolist()
        
    if df['Player'][x] == 'Casemiro':
        b_values = df.iloc[x].values.tolist()
        
a_values = a_values[1:]
b_values = b_values[1:]

values = [a_values, b_values]
```

```python
title = dict(
    title_name = 'Fabinho',
    title_color = 'red',
    subtitle_name = 'Liverpool',
    subtitle_color = 'red',
    
    title_name_2 = 'Casemiro',
    title_color_2 = 'blue',
    subtitle_name_2 = 'Real Madrid',
    subtitle_color_2 = 'blue',
    
    title_fontsize = 18, 
    subtitle_fontsize = 25
    
)

endnote = '@mrees002\ndata via FBREF / Statsbomb'
```

```python
radar = Radar()

fig,ax = radar.plot_radar(ranges = ranges, params = params, values = values,
                         radar_color = ['red', 'blue'],
                         alphas = [0.75, 0.6], title = title, endnote = endnote,
                         compare = True)
```
