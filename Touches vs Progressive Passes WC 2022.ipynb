```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Data = pd.read_csv("World Cup Possesion Stats 11_25 - Sheet1.csv")
```

```python
Touches = list(Data['Touches'])
Progressive_Passes = list(Data['Prog'])
Teams = list(Data['Squad'])
```

```python
colors = [25,20,18,21,21,23,18,15,18,24,28,19,26,15,22,22,22,30,19,27,31,32,34,33,41,23,35,33,29,29,24,34]

plt.scatter(Progressive_Passes, Touches, edgecolor = 'black', alpha = 0.7, c = colors)
plt.title('Progression Stats')
plt.xlabel('Progressive Passes recieved per 90 minutes')
plt.ylabel('Touches per 90 minutes')
cbar = plt.colorbar()
cbar.set_label('touches per progressive pass')

#plt.annotate(Teams[0], (Progressive_Passes[0], Touches[0]))
#plt.annotate(Teams[1], (Progressive_Passes[1], Touches[1]))
#plt.annotate(Teams[2], (Progressive_Passes[2], Touches[2]))
#plt.annotate(Teams[3], (Progressive_Passes[3], Touches[3]))
#plt.annotate(Teams[31], (Progressive_Passes[31], Touches[31]))
#plt.annotate(Teams[24], (Progressive_Passes[24], Touches[24]))
```
